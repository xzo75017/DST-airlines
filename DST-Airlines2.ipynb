```python
import requests 
import pandas as pd
import datetime 
import itertools

def Flight_Schedules():

    origins = ["JFK","FRA"]
    destinations = ["FRA", "CDG", "ORY"]
    
    start_date = datetime.date(2023, 4, 25)
    num_days = 4
    dates = [start_date + datetime.timedelta(days=x) for x in range(num_days)]
    dates = [d.strftime('%Y-%m-%d') for d in dates]

    
    for origin, destination, date in itertools.product(origins, destinations, dates):
        if origin == destination:
            continue
        try:

            liste_schedule_Total_Journey = []
            liste_schedule_Flight = []
            schedule = requests.get('https://api.lufthansa.com/v1/operations/schedules/{origin}/{destination}/{date}?directFlights=0'.format(origin = origin, destination = destination, date = date), headers={'Authorization': 'Bearer 3ytw9dfzypwsm7mvep8fqhuy'}).json()
            print(schedule['ScheduleResource']['Schedule'][0]['Flight'])
    #schedule_df = pd.DataFrame(schedule['ScheduleResource']['Schedule'][0]['Flight'])
            for i in range(len(schedule['ScheduleResource']['Schedule'])):
                liste_schedule_Total_Journey.append(schedule['ScheduleResource']['Schedule'][i]['TotalJourney'])
            if type(schedule['ScheduleResource']['Schedule'][i]['Flight']) is dict:
                liste_schedule_Flight.append(schedule['ScheduleResource']['Schedule'][i]['Flight']) 

            elif type(schedule['ScheduleResource']['Schedule'][i]['Flight'] is list):
                liste_schedule_Flight.append(schedule['ScheduleResource']['Schedule'][i]['Flight'][0])
                                             

        

            Schedule_Total_Journey_df = pd.DataFrame(liste_schedule_Total_Journey)
            Schedule_Flight_df = pd.DataFrame(liste_schedule_Flight)
            def AirportCode_d(Departure):
                return Departure.get('AirportCode',{})

            def ScheduledTimeLocal_d(Departure):
                return Departure.get('ScheduledTimeLocal',{}).get('DateTime',{})

            def AirportCode_a(Arrival):
                return Arrival.get('AirportCode',{})

            def ScheduledTimeLocal_a(Arrival):
                return Arrival.get('ScheduledTimeLocal',{}).get('DateTime',{})

            def Name(Arrival):
                return Arrival.get('Terminal',{}).get('Name',{})
            def AirlineID_mark(MarketingCarrier):
                return MarketingCarrier.get('AirlineID',{})

            def FlightNumber_mark(MarketingCarrier):
                return MarketingCarrier.get('FlightNumber',{})
    

    
    

            def AircraftCode(Equipment):
                return Equipment.get('AircraftCode',{})    
            def StopQuantity(Details):
                return Details.get('StopQuantity')
    
       
            def DaysOfOperation(Details):
                return Details.get('DaysOfOperation',{})
    
       
            def Effective(Details):
                return Details.get('DatePeriod',{}).get('Effective',{})    
            def Expiration(Details):
                return Details.get('DatePeriod',{}).get('Expiration',{}) 
       

            Schedule_Flight_df['AirportCode_Departure'] = Schedule_Flight_df['Departure'].apply(AirportCode_d)
            Schedule_Flight_df['ScheduledTimeLocal_Departure'] = Schedule_Flight_df['Departure'].apply(ScheduledTimeLocal_d)
            Schedule_Flight_df['ScheduledTimeLocal_Arrival'] = Schedule_Flight_df['Arrival'].apply(ScheduledTimeLocal_a)
            Schedule_Flight_df['AirportCode_Arrival'] = Schedule_Flight_df['Arrival'].apply(AirportCode_a)
            Schedule_Flight_df['Name_Terminal'] = Schedule_Flight_df['Arrival'].apply(Name)
            Schedule_Flight_df['AirlineID_Marketing_Carrier'] = Schedule_Flight_df['MarketingCarrier'].apply(AirlineID_mark)
            Schedule_Flight_df['FlightNumber_Marketing_Carrier'] = Schedule_Flight_df['MarketingCarrier'].apply(FlightNumber_mark)
    #Schedule_Flight_df['AirlineID_Operating_Carrier'] = Schedule_Flight_df['OperatingCarrier'].apply(Airlineid)
    #Schedule_Flight_df['FlightNumber_Operating_Carrier'] = Schedule_Flight_df['OperatingCarrier'].apply(FlightNumber_op)
    #Schedule_Flight_df['AirlineID_Operating_Carrier'] = Schedule_Flight_df['OperatingCarrier'].apply(FlightNumber)
            Schedule_Flight_df['AircraftCode'] = Schedule_Flight_df['Equipment'].apply(AircraftCode)
            Schedule_Flight_df['StopQuantity'] = Schedule_Flight_df['Details'].apply(StopQuantity)
            Schedule_Flight_df['DaysOfOperation'] = Schedule_Flight_df['Details'].apply(DaysOfOperation)
            Schedule_Flight_df['Effective'] = Schedule_Flight_df['Details'].apply(Effective)
            Schedule_Flight_df['Expiration'] = Schedule_Flight_df['Details'].apply(Expiration)

            Schedule_Flight_df = Schedule_Flight_df.drop(['Departure', 'Arrival','MarketingCarrier', 'Equipment','Details'], axis = 1)
            Schedule_Flight_df.head()
        
        except: 

            print(f"La requête API a échoué pour la combinaison : {origin}/{destination} à la date : {date}")
            continue
        return Schedule_Flight_df, Schedule_Total_Journey_df
```
