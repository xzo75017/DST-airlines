```python
import pandas as pd
import requests

def arrival():
    list_arrival = []
    data_arrival_flight = requests.get('https://api.lufthansa.com/v1/operations/customerflightinformation/arrivals/FRA/2023-04-28T15:00' , headers={'Authorization': 'Bearer czdgawxxync46475qcdsg7qr'}).json()
    #Visualisation du fichier json on remarque que la liste est à data_arrival_flight['FlightInformation']['Flights']['Flight'] + visualisation des clés des dictionnaires
        #print(data_arrival_flight['FlightInformation']['Flights']['Flight'][0].keys())
    #Boucle sur chaque élément de la liste
    for i in range(len(data_arrival_flight['FlightInformation']['Flights']['Flight'])):
#Visualisation des différentes variables contenues dans la liste
        print(data_arrival_flight['FlightInformation']['Flights']['Flight'][0])
        #liste_Departure.append(data_arrival_flight['FlightInformation']['Flights']['Flight'][i]['Departure'])
        #liste_departure_schedule.append(data_arrival_flight['FlightInformation']['Flights']['Flight'][i]['Departure']['Scheduled'])
        #liste_departure_actual.append(data_arrival_flight['FlightInformation']['Flights']['Flight'][i]['Arrival']['Actual'])
        list_arrival.append(data_arrival_flight['FlightInformation']['Flights']['Flight'][i])
        #liste_terminal.append(data_arrival_flight['FlightInformation']['Flights']['Flight'][i]['Departure']['Terminal'])
        #liste_arrival_status.append(data_arrival_flight['FlightInformation']['Flights']['Flight'][i]['Departure']['Status'])
    #Departure = pd.DataFrame(liste_Departure)
    #Departure_Scheduled = pd.DataFrame(liste_departure_schedule)
    #Arrival_Actual  = pd.DataFrame(liste_arrival_actual)
    Arrival = pd.DataFrame(list_arrival)


    def AirportCode_de(Departure):
        return Departure.get('AirportCode',{})

    def Scheduled_date_de(Departure):
        return Departure.get('Scheduled',{}).get('Date',{})
    
    def Scheduled_time_de(Departure):
        return Departure.get('Scheduled',{}).get('Time',{})

    def Scheduled_date_ae(Arrival):
        return Arrival.get('Scheduled',{}).get('Date',{})
    
    def Scheduled_time_ae(Arrival):
        return Arrival.get('Scheduled',{}).get('Time',{})

    def Name(Arrival):
        return Arrival.get('Terminal',{}).get('Name',{})
    
    def AirportCode_ae(Arrival):
        return Arrival.get('AirportCode',{})
    
    def estimated_date_ae(Arrival):
        return Arrival.get('Estimated',{}).get('Time',{})
    
    def estimated_time_ae(Arrival):
        return Arrival.get('Estimated',{}).get('Date',{})
    
    def actual_time_ae(Arrival):
        return Arrival.get('Actual',{}).get('Time',{})
    
    def actual_date_ae(Arrival):
        return Arrival.get('Actual',{}).get('Date',{})
    
    def actual_time_de(Departure):
        return Departure.get('Actual',{}).get('Time',{})
    
    def actual_date_de(Departure):
        return Departure.get('Actual',{}).get('Date',{})
    
       
    for i in range(len(data_arrival_flight['FlightInformation']['Flights']['Flight'][0]['MarketingCarrierList'])):
        def AirlineID_mark(MarketingCarrierList):
            return MarketingCarrierList.get('MarketingCarrier',{})[i]

    def FlightNumber(MarketingCarrierList):
        return MarketingCarrierList.get('MarketingCarrier',{})
    
    def AirlineID(OperatingCarrier):
        return OperatingCarrier.get('AirlineID',{})
    
    def FlightNumber(OperatingCarrier):
        return OperatingCarrier.get('FlightNumber',{})
   
    def AircraftCode(Equipment):
        return Equipment.get('AircraftCode',{}) 
       
    def Status_Code_d(Departure):
        return Departure.get('Status',{}).get('Code',{})
    
    def Status_Description_d(Departure):
        return Departure.get('Status',{}).get('Description',{})
    

    def Status_Code_a(Arrival):
        return Arrival.get('Status',{}).get('Code',{})
    
    def Status_Description_a(Arrival):
        return Arrival.get('Status',{}).get('Description',{})
    

    
    #def AirlineID_Marketing_Carrier_(AirlineID_Marketing_Carrier):
    #    return AirlineID_Marketing_Carrier.get('AirlineID')
    
    #def Flight_Number_Marketing_Carrier_(AirlineID_Marketing_Carrier):
    #    return AirlineID_Marketing_Carrier.get('FlightNumber')  
        #Terminal_Departure = pd.DataFrame(liste_terminal)

    Arrival['AirportCode_Departure'] = Arrival['Departure'].apply(AirportCode_de)
    Arrival['ScheduledDateLocal_Departure'] = Arrival['Departure'].apply(Scheduled_date_de)
    Arrival['ScheduledTimeLocal_Departure'] = Arrival['Departure'].apply(Scheduled_time_de)
    Arrival['ScheduledDateLocal_Arrival'] = Arrival['Arrival'].apply(Scheduled_date_ae)
    Arrival['ScheduledTimeLocal_Arrival'] = Arrival['Arrival'].apply(Scheduled_time_ae)
    Arrival['AirportCode_Arrival'] = Arrival['Arrival'].apply(AirportCode_ae)
    Arrival['Name_Terminal'] = Arrival['Arrival'].apply(Name)
    Arrival['Estimated_date_arrival'] = Arrival['Arrival'].apply(estimated_date_ae)
    Arrival['Estimated_time_arrival'] = Arrival['Arrival'].apply(estimated_time_ae)
    Arrival['Actual_date_arrival'] = Arrival['Arrival'].apply(actual_date_ae)
    Arrival['Actual_time_arrival'] = Arrival['Arrival'].apply(actual_time_ae)
    Arrival['Actual_date_departure'] = Arrival['Departure'].apply(actual_date_de)
    Arrival['Actual_time_departure'] = Arrival['Departure'].apply(actual_time_de)
    #Departure_departure['AirlineID_Marketing_Carrier'] = Departure_departure['MarketingCarrierList'].apply(AirlineID_mark)
    #Departure_departure['AirlineID_Marketing_Carrier_'] = Departure_departure['AirlineID_Marketing_Carrier'].apply(AirlineID_Marketing_Carrier_)
    #Departure_departure['FlightNumber_Marketing_Carrier'] = Departure_departure['AirlineID_Marketing_Carrier'].apply(Flight_Number_Marketing_Carrier_)
    Arrival['AirlineID_Operating_Carrier'] = Arrival['OperatingCarrier'].apply(AirlineID)
    Arrival['FlightNumber_Operating_Carrier'] = Arrival['OperatingCarrier'].apply(FlightNumber)
    Arrival['AircraftCode'] = Arrival['Equipment'].apply(AircraftCode)
    Arrival['Status_Code_Departure'] = Arrival['Departure'].apply(Status_Code_d)
    Arrival['Status_Description_Departure'] = Arrival['Departure'].apply(Status_Description_d)
    Arrival['Status_Code_Arrival'] = Arrival['Arrival'].apply(Status_Code_a)
    Arrival['Status_Description_Arrival'] = Arrival['Arrival'].apply(Status_Description_a)
    Arrival = Arrival.drop(['Departure',	'Arrival',	'OperatingCarrier',	'Equipment','Status', 'MarketingCarrierList'], axis = 1)

    #Terminal = pd.DataFrame(liste_terminal)
        #Terminal = pd.DataFrame(liste_terminal)
    return Arrival
```
