```python
import pandas as pd
import requests
from datetime import datetime, timedelta
def departure():    
    liste_Departure = []
    #Requête via la protocole HTTP
    airports_codes= ["FRA","BER"]
    d = '2023-04-27T09:30'

    dt = datetime.strptime(d, '%Y-%m-%dT%H:%M')
    
    while dt< datetime.now() :
        g = dt+timedelta(days = 1)
        dt = g
        g = g.strftime("%Y-%m-%dT%H:%M")

        print(g)
        for airport_code in airports_codes:
            data_Departure_flight = requests.get(f'https://api.lufthansa.com/v1/operations/customerflightinformation/departures/{airport_code}/{g}' , headers={'Authorization': 'Bearer czdgawxxync46475qcdsg7qr'}).json()
        #Visualisation du fichier json on remarque que la liste est à data_arrival_flight['FlightInformation']['Flights']['Flight'] + visualisation des clés des dictionnaires
            print(data_Departure_flight['FlightInformation']['Flights']['Flight'][0])
        #Boucle sur chaque élément de la liste
            for i in range(len(data_Departure_flight['FlightInformation']['Flights']['Flight'])):
#Visualisation des différentes variables contenues dans la liste
        #print(data_Departure_flight['FlightInformation']['Flights']['Flight'][3]['MarketingCarrierList'])
                liste_Departure.append(data_Departure_flight['FlightInformation']['Flights']['Flight'][i])

        #liste_marketing_carrier(data_Departure_flight['FlightInformation']['Flights']['Flight'][i]['OperatingCarrier'])
    Departure_departure= pd.DataFrame(liste_Departure)

    def AirportCode_de(Departure):
        return Departure.get('AirportCode',{})

    def Scheduled_date_de(Departure):
        return Departure.get('Scheduled',{}).get('Date',{})
    
    def Scheduled_time_de(Departure):
        return Departure.get('Scheduled',{}).get('Time',{})

    def Scheduled_date_ae(Arrival):
        return Arrival.get('Scheduled',{}).get('Date',{})
    
    def Scheduled_time_ae(Arrival):
        return Arrival.get('Scheduled',{}).get('Time',{})

    def Name(Arrival):
        return Arrival.get('Terminal',{}).get('Name',{})
    
    def AirportCode_ae(Arrival):
        return Arrival.get('AirportCode',{})
    
    def estimated_date_ae(Arrival):
        return Arrival.get('Estimated',{}).get('Time',{})
    
    def estimated_time_ae(Arrival):
        return Arrival.get('Estimated',{}).get('Date',{})
    
    def actual_time_ae(Arrival):
        return Arrival.get('Actual',{}).get('Time',{})
    
    def actual_date_ae(Arrival):
        return Arrival.get('Actual',{}).get('Date',{})
    
    def actual_time_de(Departure):
        return Departure.get('Actual',{}).get('Time',{})
    
    def actual_date_de(Departure):
        return Departure.get('Actual',{}).get('Date',{})
    
       
    #for i in range(len(data_Departure_flight['FlightInformation']['Flights']['Flight'][0]['MarketingCarrierList'])):
    #    def AirlineID_mark(MarketingCarrierList):
    #        return MarketingCarrierList.get('MarketingCarrier',{})[i]

    def FlightNumber(MarketingCarrierList):
        return MarketingCarrierList.get('MarketingCarrier',{})
    
    def AirlineID(OperatingCarrier):
        return OperatingCarrier.get('AirlineID',{})
    
    def FlightNumber(OperatingCarrier):
        return OperatingCarrier.get('FlightNumber',{})
   
    def AircraftCode(Equipment):
        return Equipment.get('AircraftCode',{}) 
       
    def Status_Code_d(Departure):
        return Departure.get('Status',{}).get('Code',{})
    
    def Status_Description_d(Departure):
        return Departure.get('Status',{}).get('Description',{})
    

    def Status_Code_a(Arrival):
        return Arrival.get('Status',{}).get('Code',{})
    
    def Status_Description_a(Arrival):
        return Arrival.get('Status',{}).get('Description',{})
    

    
    def AirlineID_Marketing_Carrier_(AirlineID_Marketing_Carrier):
        return AirlineID_Marketing_Carrier.get('AirlineID')
    
    def Flight_Number_Marketing_Carrier_(AirlineID_Marketing_Carrier):
        return AirlineID_Marketing_Carrier.get('FlightNumber')  
        #Terminal_Departure = pd.DataFrame(liste_terminal)

    Departure_departure['AirportCode_Departure'] = Departure_departure['Departure'].apply(AirportCode_de)
    Departure_departure['ScheduledDateLocal_Departure'] = Departure_departure['Departure'].apply(Scheduled_date_de)
    Departure_departure['ScheduledTimeLocal_Departure'] = Departure_departure['Departure'].apply(Scheduled_time_de)
    Departure_departure['ScheduledDateLocal_Arrival'] = Departure_departure['Arrival'].apply(Scheduled_date_ae)
    Departure_departure['ScheduledTimeLocal_Arrival'] = Departure_departure['Arrival'].apply(Scheduled_time_ae)
    Departure_departure['AirportCode_Arrival'] = Departure_departure['Arrival'].apply(AirportCode_ae)
    Departure_departure['Name_Terminal'] = Departure_departure['Arrival'].apply(Name)
    Departure_departure['Estimated_date_arrival'] = Departure_departure['Arrival'].apply(estimated_date_ae)
    Departure_departure['Estimated_time_arrival'] = Departure_departure['Arrival'].apply(estimated_time_ae)
    Departure_departure['Actual_date_arrival'] = Departure_departure['Arrival'].apply(actual_date_ae)
    Departure_departure['Actual_time_arrival'] = Departure_departure['Arrival'].apply(actual_time_ae)
    Departure_departure['Actual_date_departure'] = Departure_departure['Departure'].apply(actual_date_de)
    Departure_departure['Actual_time_departure'] = Departure_departure['Departure'].apply(actual_time_de)
    #Departure_departure['AirlineID_Marketing_Carrier'] = Departure_departure['MarketingCarrierList'].apply(AirlineID_mark)
    #Departure_departure['AirlineID_Marketing_Carrier_'] = Departure_departure['AirlineID_Marketing_Carrier'].apply(AirlineID_Marketing_Carrier_)
    #Departure_departure['FlightNumber_Marketing_Carrier'] = Departure_departure['AirlineID_Marketing_Carrier'].apply(Flight_Number_Marketing_Carrier_)
    Departure_departure['AirlineID_Operating_Carrier'] = Departure_departure['OperatingCarrier'].apply(AirlineID)
    Departure_departure['FlightNumber_Operating_Carrier'] = Departure_departure['OperatingCarrier'].apply(FlightNumber)
    Departure_departure['AircraftCode'] = Departure_departure['Equipment'].apply(AircraftCode)
    Departure_departure['Status_Code_Departure'] = Departure_departure['Departure'].apply(Status_Code_d)
    Departure_departure['Status_Description_Departure'] = Departure_departure['Departure'].apply(Status_Description_d)
    Departure_departure['Status_Code_Arrival'] = Departure_departure['Arrival'].apply(Status_Code_a)
    Departure_departure['Status_Description_Arrival'] = Departure_departure['Arrival'].apply(Status_Description_a)
    Departure_departure = Departure_departure.drop(['Departure',	'Arrival',	'OperatingCarrier',	'Equipment','Status','MarketingCarrierList' ], axis = 1)
    return Departure_departure
```
